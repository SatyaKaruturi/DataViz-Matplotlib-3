# Mission 3

Modifie le code ci-dessous dans une nouvelle cellule pour que :
- Les valeurs pour hommes soient à droite et celles pour femmes à gauche
- Modifier la couleur des colonnes
- Diminuer l'épaisseur des paires de baton

```
import matplotlib.pyplot as plt
import numpy as np


labels = ['G1', 'G2', 'G3', 'G4', 'G5']
men_means = [20, 34, 30, 35, 27] 
women_means = [25, 32, 34, 20, 25]

x = np.arange(len(labels)) 
width = 0.35  

fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, men_means, width, label='Men')
rects2 = ax.bar(x + width/2, women_means, width, label='Women')

ax.set_ylabel('Scores')
ax.set_title('Scores by group and gender')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()


def autolabel(rects):
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3), 
                    textcoords="offset points",
                    ha='center', va='bottom')


autolabel(rects1)
autolabel(rects2)

fig.tight_layout()

plt.show()
```

# Mission 4

Commente le code ci-dessous pour l'expliquer à un collège

```
import matplotlib
import numpy as np
import matplotlib.pyplot as plt #pour fair les plot on import matplotlib, numpy pour la calucaltions

np.random.seed(19680801) #Cette méthode est appelée lorsque L'état aléatoire est initialisé. Il peut être appelé à nouveau pour relancer le générateur

mu = 100  
sigma = 15  
x = mu + sigma * np.random.randn(437) #alcule the valuer pour x

num_bins = 50 #a plage de valeurs - c'est-à diviser l'ensemble de la gamme de valeurs dans une série d'intervalles et puis compter le nombre de valeurs se situent dans chaque intervalle

fig, ax = plt.subplots() #plot le subplot

n, bins, patches = ax.hist(x, num_bins, density=1)  # creé le histogram

y = ((1 / (np.sqrt(2 * np.pi) * sigma)) *
     np.exp(-0.5 * (1 / sigma * (bins - mu))**2)) # calcule the valuer Y pour plot le curve
ax.plot(bins, y, '--') #plot le curve sur la histogram, define quel type line on aime voire '--'
ax.set_xlabel('Smarts') #label x axis
ax.set_ylabel('Probability density') #label y axis
ax.set_title(r'Histogram of IQ: $\mu=100$, $\sigma=15$') ##label title

fig.tight_layout() #tight_layout ajuste automatiquement les sous-lots de manière à ce que le ou les sous-lots s'insèrent dans la zone des figures.
plt.show() #montre le plot
```
